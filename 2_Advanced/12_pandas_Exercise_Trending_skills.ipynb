```python

# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short']=='Data Analyst'].copy()
```

```python
df_DA['job_posted_date'] = pd.to_datetime(df_DA['job_posted_date'])

df_DA['job_posted_month'] = df_DA['job_posted_date'].dt.strftime('%b')

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month', 
                                        columns='job_skills', 
                                        aggfunc='size', 
                                        fill_value=0)

df_DA_pivot = df_DA_pivot.reindex(month_order)


df_DA_pivot = df_DA_pivot[df_DA_pivot.sum().sort_values(ascending=False).index]

```

```python
df_DA['job_posted_month_no']=df_DA['job_posted_date'].dt.month
df_DA_explode= df_DA.explode('job_skills')
df_DA_pivot= df_DA_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)
```

```python
df_DA_pivot.loc['Total']=df_DA_pivot.sum()
df_DA_pivot=df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot=df_DA_pivot.drop('Total')
df_DA_pivot
```

```python
df_DA_pivot.iloc[:,:5].plot(kind='line')

plt.title('Top 5 Skills for Data Analyst per Month')
plt.xlabel('')
plt.ylabel('Count')
plt.show
```
